```python
import pandas as pd

import pandas as pd

# Load your data

# Initialize dictionaries to store the counts of zeroes and negative values
zero_counts = {}
negative_counts = {}

# Iterate over each column in the DataFrame
for column in df.columns:
    if pd.api.types.is_numeric_dtype(df[column]):  # Only check numeric columns
        zero_counts[column] = (df[column] == 0).sum()
        negative_counts[column] = (df[column] < 0).sum()

# Convert the dictionaries to DataFrames for better display
zero_counts_df = pd.DataFrame(list(zero_counts.items()), columns=['Column', 'Zero Count'])
negative_counts_df = pd.DataFrame(list(negative_counts.items()), columns=['Column', 'Negative Count'])

# Display the results
print("Zero Counts in Each Column:")
print(zero_counts_df)

print("\nNegative Counts in Each Column:")
print(negative_counts_df)
```

```python

null_values = df.isnull().sum()
print("Null values in each column:\n", null_values)

def impute_missing_values(df):
    for column in df.columns:
        if df[column].isnull().sum() > 0:
            if df[column].dtype == 'object':
                # For categorical variables, use mode
                df[column].fillna(df[column].mode()[0], inplace=True)
            else:
                # For numerical variables, use median
                df[column].fillna(df[column].median(), inplace=True)

# Impute missing values
impute_missing_values(df)

# Verify if there are any remaining null values
null_values_after = df.isnull().sum()
print("Null values in each column after imputation:\n", null_values_after)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Sample data frame based on your provided columns

# List of numerical fields for outlier removal
fields_for_outlier_removal = [
    'Income (USD)',
    'Loan Amount Request (USD)',
    'Current Loan Expenses (USD)',
    'Credit Score',
    'Age',
    'Property Age',
    'Property Price',
    'Loan Sanction Amount (USD)'
]

# Function to detect and remove outliers using IQR method
def remove_outliers_iqr(data, fields):
    cleaned_data = data.copy()
    outliers_before = {}
    outliers_after = {}
    
    for field in fields:
        Q1 = cleaned_data[field].quantile(0.25)
        Q3 = cleaned_data[field].quantile(0.75)
        IQR = Q3 - Q1
        lower_limit = Q1 - 1.5 * IQR
        upper_limit = Q3 + 1.5 * IQR
        
        outliers_before[field] = len(cleaned_data[(cleaned_data[field] < lower_limit) | (cleaned_data[field] > upper_limit)])
        cleaned_data = cleaned_data[(cleaned_data[field] >= lower_limit) & (cleaned_data[field] <= upper_limit)]
        
        outliers_after[field] = len(cleaned_data[(cleaned_data[field] < lower_limit) | (cleaned_data[field] > upper_limit)])
    
    return cleaned_data, outliers_before, outliers_after

# Function to detect and remove outliers using Z-score method
def remove_outliers_zscore(data, fields):
    cleaned_data = data.copy()
    outliers_before = {}
    outliers_after = {}
    
    for field in fields:
        z_scores = np.abs(stats.zscore(cleaned_data[field]))
        outliers_before[field] = len(cleaned_data[np.abs(z_scores) > 3])
        cleaned_data = cleaned_data[(np.abs(z_scores) < 3)]
        
        z_scores = np.abs(stats.zscore(cleaned_data[field]))
        outliers_after[field] = len(cleaned_data[np.abs(z_scores) > 3])
    
    return cleaned_data, outliers_before, outliers_after

# Remove outliers using IQR method
df_cleaned_iqr, outliers_iqr_before, outliers_iqr_after = remove_outliers_iqr(df, fields_for_outlier_removal)

# Remove outliers using Z-score method
df_cleaned_zscore, outliers_zscore_before, outliers_zscore_after = remove_outliers_zscore(df, fields_for_outlier_removal)

# Plotting boxplots before and after outlier removal using IQR method
plt.figure(figsize=(16, 10))

for i, field in enumerate(fields_for_outlier_removal):
    plt.subplot(2, len(fields_for_outlier_removal)//2, i+1)
    sns.boxplot(x=df[field], color='skyblue')
    plt.title(f'Before IQR: {field}')
    plt.tight_layout()

plt.figure(figsize=(16, 10))

for i, field in enumerate(fields_for_outlier_removal):
    plt.subplot(2, len(fields_for_outlier_removal)//2, i+1)
    sns.boxplot(x=df_cleaned_iqr[field], color='lightgreen')
    plt.title(f'After IQR: {field}')
    plt.tight_layout()

plt.show()

# Plotting boxplots before and after outlier removal using Z-score method
plt.figure(figsize=(16, 10))

for i, field in enumerate(fields_for_outlier_removal):
    plt.subplot(2, len(fields_for_outlier_removal)//2, i+1)
    sns.boxplot(x=df[field], color='skyblue')
    plt.title(f'Before Z-score: {field}')
    plt.tight_layout()

plt.figure(figsize=(16, 10))

for i, field in enumerate(fields_for_outlier_removal):
    plt.subplot(2, len(fields_for_outlier_removal)//2, i+1)
    sns.boxplot(x=df_cleaned_zscore[field], color='lightgreen')
    plt.title(f'After Z-score: {field}')
    plt.tight_layout()

plt.show()

# Print number of outliers detected and removed
print("\nOutliers detected and removed using IQR method:")
print("Before removal:")
print(outliers_iqr_before)
print("\nAfter removal:")
print(outliers_iqr_after)

print("\nOutliers detected and removed using Z-score method:")
print("Before removal:")
print(outliers_zscore_before)
print("\nAfter removal:")
print(outliers_zscore_after)
```
